# Basic EDA

```python
def eda(df, name="data"): 
    print(f"\n EDA FOR: {name} \n")
    
    
    print(f"Shape: {df.shape}")
    
    print("\n Data Types: ")
    print(df.dtypes.value_counts())
    
    print("\n Head: ")
    print(df.head())
    
    print("\n Info: ")
    print(df.info())
    
    print("\n Describe: ")
    print(df.describe())
    
    print("\n Features Unique & Missing Val: ")
    features = pd.DataFrame(columns=['Features','Number of unique values','Number of nulls'])
    for i, feat in enumerate(df.columns):
        features.loc[i] = [feat, df[feat].nunique(), df[feat].isnull().sum()]
    print(features)
    
    print("\n Num Ft vs Cat Ft: ")
    num_ft = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
    cat_ft = df.select_dtypes(include=["object", "bool"]).columns.tolist()
    print(f"\n Numerical ft: {num_ft} \n Total = {len(num_ft)}")
    print(f"\n Categorical ft: {cat_ft} \n Total = {len(cat_ft)}")

    print("\n Cat ft: ")
    for col in cat_ft:
        unique_values = df[col].nunique()
        print(f"{col}: {unique_values} unique values")
    
    
    
```

```python
df = pd.read_csv("""data file""")
eda(df, name="data name")
```

# For ERD

```python
def col_info(datasets): 
    dataset_col = {}
    for name, df in datasets.items():
        dataset_col[name] = set(df.columns)
    
    col_dict = {}
    
    for name, col in dataset_col.items():
        col_series = pd.Series(list(col))
        col_dict[name] = col_series
    
    features_df = pd.DataFrame(col_dict)
    
    overlaps = []
    dataset_names = list(datasets.keys())
    
    for x in range(len(dataset_names)):
        for y in range(x+1, len(dataset_names)):
            d1 = dataset_names[x]
            d2 = dataset_names[y]
            
            shared = dataset_col[d1].intersection(dataset_col[d2])
            
            if shared:
                overlaps.append({
                    "Dataset 1": d1, 
                    "Dataset 2": d2, 
                    "Num of shared col": len(shared), 
                    "Shared col": shared
                })
                
    overlaps_df = pd.DataFrame(overlaps)
    
    return features_df, overlaps_df
```

```python
datasets = {
    """columns"""
}
```

```python
features_df, overlaps_df = col_info(datasets)

features_df.to_excel("eda_feature_lists.xlsx", index=False)
overlaps_df.to_excel("eda_overlaps.xlsx", index=False)
```
