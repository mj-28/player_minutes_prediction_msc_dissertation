# 3. Building Model

## 3.1 Selecting Features

```python
selector = SelectKBest(score_func=f_regression, k=90)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)
```

## 3.2 Grid Search

```python
param_grid = {
    'n_estimators': [100, 300],
    'max_depth': [None, 10, 20],
    'max_features': ['sqrt', 0.5],
    'min_samples_leaf': [1, 3, 5]
}

rfr = RandomForestRegressor(random_state=0)

grid_search = GridSearchCV(
    estimator=rfr,
    param_grid=param_grid,
    cv=5,
    scoring='r2'
)

grid_search.fit(X_train_selected, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best CV R2", grid_search.best_score_)

best_model = grid_search.best_estimator_
```

#### Test & Evaluate
- skip this if you want to do the interpretation methods

```python
rfr_y_pred = best_model.predict(X_test_selected)
```

```python
score = best_model.score(X_test_selected, y_test)
mse = mean_squared_error(y_test, rfr_y_pred)
mae = mean_absolute_error(y_test, rfr_y_pred)
r2 = r2_score(y_test, rfr_y_pred)
rmse = np.sqrt(mse)

print("Model Evaluation Metrics:")
print(f"Score: {score:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R-squared (R²): {r2:.4f}")
```

```python
#Actual vs predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=rfr_y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label="Perfect Fit")
plt.title("Actual vs Predicted Values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.show()
```

```python
#Residual histogram
residuals = y_test - rfr_y_pred

plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, bins=30, color="blue")
plt.title("Residual Distribution")
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.axvline(0, color="red", linestyle="--", label="No Error")
plt.legend()
plt.show()
```
